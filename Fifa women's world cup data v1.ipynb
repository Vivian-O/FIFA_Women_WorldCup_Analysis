# Actual work

```python
#actual work cell
#details of all countries in the dataset
allSquadsStages = df[['squad','year','matches_played']]
allSquadsStages
```

```python
#actual work cell
#getting the stages of all the countries in each year that they participated
#a null value indicates they didn't participate for that year
yearStages_wide = pd.pivot(allSquadsStages, index=['squad'], columns = 'year',values = 'matches_played').reset_index()#Reshape from long to wide
yearStages_wide
```

```python
#yearStages_wide.to_csv("country stages.csv")
yearStages_wide.info()
```

```python
#actual work cell
#Average stage of the all the countries in the dataset
yearStages_wide['Average stage'] = round(yearStages_wide.mean(axis=1),0)
yearStages_wide
```

```python
#grouping squads into average stages
groupStages = yearStages_wide[yearStages_wide['Average stage'] == 3]
roundOf16 = yearStages_wide[yearStages_wide['Average stage'] == 4]
quarterFinals = yearStages_wide[yearStages_wide['Average stage'] == 5]
semiAndFinals = yearStages_wide[yearStages_wide['Average stage'] >= 6]
```

```python
#seeing squads that have their average as the group stage
groupStagesList = list(groupStages['squad'])
groupStagesList
```

```python
roundOf16 = list(roundOf16['squad'])
quarterFinals = list(quarterFinals['squad'])
semiAndFinals = list(semiAndFinals['squad'])
```

```python
roundOf16
```

```python
# Using the lists of squads and their average stages, getting all the
# records of those squads from the original dataframe and storing
#them in their own dataframes
groupStagesdf = df[df['squad'].isin(groupStagesList)]
roundOf16df = df[df['squad'].isin(roundOf16)]
quarterFinalsdf = df[df['squad'].isin(quarterFinals)]
semiAndFinalsdf = df[df['squad'].isin(semiAndFinals)]
```

```python
#looking at the records of the countries whose average stage is the Round of 16
roundOf16df
```

```python
#dropping the year, id and starts columns as I am going to get the mean
# of the squads stats in each column i.e for all of Australia's appearances,
# what was its average number of goals
groupStagesdf.drop(['year','id','starts'], axis=1,inplace=True)
roundOf16df.drop(['year','id','starts'], axis=1,inplace=True)
quarterFinalsdf.drop(['year','id','starts'], axis=1,inplace=True)
semiAndFinalsdf.drop(['year','id','starts'], axis=1,inplace=True) 
```

```python
roundOf16df
```

```python
#getting the mean
# of the squads stats in each column
groupStagesAverages = groupStagesdf.groupby('squad').mean()
roundOf16Averages = roundOf16df.groupby('squad').mean()
quarterFinalsAverages = quarterFinalsdf.groupby('squad').mean()
semiAndFinalsAverages = semiAndFinalsdf.groupby('squad').mean()
```

```python
#library to get the continent provided country name
import pycountry_convert as pc

#function to get the continent provided country name
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

#calling the function on the required dataframe to get continents
yearStages_wide['continent'] = yearStages_wide['squad'].map(country_to_continent)
```
